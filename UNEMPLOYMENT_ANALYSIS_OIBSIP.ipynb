```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm

# Step 1: Data Collection
data = pd.read_csv('Unemployment in India.csv')

# Step 2: Data Preprocessing
# Assuming the data contains columns "Date" and "UnemploymentRate"
data[' Date'] = pd.to_datetime(data[' Date'])
data.set_index(' Date', inplace=True)

# Step 3: Data Visualization
plt.figure(figsize=(12, 6))
sns.lineplot(data=data[' Estimated Unemployment Rate (%)'], linewidth=2, color='b')
plt.title('Unemployment Rate Over Time')
plt.xlabel(' Date')
plt.ylabel(' Estimated Unemployment Rate (%)')
plt.grid(True)
plt.show()

# Step 4: Statistical Analysis
# Assuming you want to compare pre-pandemic and pandemic periods (e.g., 2019 and 2020)
pre_pandemic_data = data[' Estimated Unemployment Rate (%)']['2019']
pandemic_data = data[' Estimated Unemployment Rate (%)']['2020']

# Step 5: Time Series Analysis (Optional)
# Assuming you have enough historical data, you can perform time series forecasting
model = sm.tsa.SARIMAX(data[' Estimated Unemployment Rate (%)'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
results = model.fit()
forecast = results.forecast(steps=12)  # Forecasting 12 months ahead

plt.figure(figsize=(12, 6))
sns.lineplot(data=data[' Estimated Unemployment Rate (%)'], linewidth=2, color='b', label='Observed')
sns.lineplot(x=forecast.index, y=forecast, linewidth=2, color='r', label='Forecast')
plt.title('Unemployment Rate Forecast')
plt.xlabel('Date')
plt.ylabel(' Estimated Unemployment Rate (%)')
plt.grid(True)
plt.legend()
plt.show()
```
